```python
df01 = pd.read_csv('c:/Users/chito/Downloads/facebook/1684.edges.txt', sep=' ', header=None, names=['node1', 'node2'])
```

```python
print(df01)
```

```python
max_node_id = max(df01['node1'].max(), df01['node2'].max())
edge_list = []
for index, row in df01.iterrows():
    node1, node2 = int(row['node1']) - 1, int(row['node2']) - 1  # Subtract 1 for 0-based indexing
    edge_list.append((row['node1'],row['node2']))
```
